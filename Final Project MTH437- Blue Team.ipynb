# Algorithm of Linear Shooting Method.

1)The Boundary-value problem y" = 4(y-x), 0<=x<=1, Y(0)=0, Y(1)=2, the solution y(x) = e^2(e^4-1)^-1(e^2x-e^-2x)+x. Use the linear shooting method to approximate the solution and compare the results to the actual solution.

Importing Necessary Libraries to Solve the problem.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import math
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

Converting the given equation into -y"+p(x)y'+q(x)y+r(x) = 0 and Writng the variables as below

```python
#Step 1
n=2
y_a = 0
y_b = 2
u1 = [y_a]
u2 = [0]
v1 = [0]
v2 = [1]
a = 0
b = 1
h = (b-a)/n
p  = lambda x : 0
q  = lambda x : 4
r  = lambda x : -4*x
```

#Building the function same as above using Ranga- kutta Method.

```python
def u_v_val(n):
    
    for i in range(n):
        
        
      #step - 2 :
    
        x = a+(i*h)
        print("x :",x)
        k1_1 = h*(u2[i])
        print("k1_1 : ",k1_1)
        k1_2 = h*((p(x)*u2[i])+(q(x)*u1[i])+r(x))
        print("k1_2 : ",k1_2)
        k2_1 = h*(u2[i]+(0.5*(k1_2)))
        print("k2_1 : ",k2_1)
        k2_2 = h*((p(x+(0.5*h))*(u2[i]+(0.5*k1_2)))+(q(x+(0.5*h))*(u1[i]+(0.5*k1_1)))+r(x+(0.5*h)))
        print("k2_2 : ",k2_2)
        k3_1 = h*(u2[i]+(0.5*(k2_2)))
        print("k3_1 : ",k3_1)
        k3_2 = h*((p(x+(0.5*h))*(u2[i]+(0.5*k2_2)))+(q(x+(0.5*h))*(u1[i]+(0.5*k2_1)))+r(x+(0.5*h)))
        print("k3_2 : ",k3_2)
        k4_1 = h*(u2[i]+(k3_2))
        print("k4_1 : ",k4_1)
        k4_2 = h*((p(x+(h))*(u2[i]+(k3_2)))+(q(x+(h))*(u1[i]+(k3_1)))+r(x+(h)))
        print("k4_2 : ",k4_2)
        u1.append(u1[i]+((1/6)*(k1_1+(2*k2_1)+(2*k3_1)+k4_1)))
        print(u1[i+1])
        u2.append(u2[i]+((1/6)*(k1_2+(2*k2_2)+(2*k3_2)+k4_2)))
        print(u2[i+1])
        print("-"*100)

        k1_1_prime = h*(v2[i])
        print("k1_1' : ",k1_1_prime)
        k1_2_prime = h*((p(x)*v2[i])+(q(x)*v1[i]))
        print("k1_2' : ",k1_2_prime)
        k2_1_prime = h*(v2[i]+(0.5*(k1_2_prime)))
        print("k2_1' : ",k2_1_prime)
        k2_2_prime = h*((p(x+(0.5*h))*(v2[i]+(0.5*k1_2_prime)))+(q(x+(0.5*h))*(v1[i]+(0.5*k1_1_prime))))
        print("k2_2' : ",k2_2_prime)
        k3_1_prime = h*(v2[i]+(0.5*(k2_2_prime)))
        print("k3_1' : ",k3_1_prime)
        k3_2_prime = h*((p(x+(0.5*h))*(v2[i]+(0.5*k2_2_prime)))+(q(x+(0.5*h))*(v1[i]+(0.5*k2_1_prime))))
        print("k3_2' : ",k3_2_prime)
        k4_1_prime = h*(v2[i]+(k3_2_prime))
        print("k4_1' : ",k4_1_prime)
        k4_2_prime = h*((p(x+(h))*(v2[i]+(k3_2_prime)))+(q(x+(h))*(v1[i]+(k3_1_prime))))
        print("k1_2' : ",k4_2_prime)
        v1.append(v1[i]+((1/6)*(k1_1_prime+(2*k2_1_prime)+(2*k3_1_prime)+k4_1_prime)))
        print(v1[i+1])
        v2.append(v2[i]+((1/6)*(k1_2_prime+(2*k2_2_prime)+(2*k3_2_prime)+k4_2_prime)))
        print(v2[i+1])
        print("-"*100)
```

```python
u_v_val(2)
```

```python
#Step -3
w1_0 = y_a
print("w1_0: ",w1_0)
w2_0 = (y_b-u1[n])/(v1[n])
print("w2_0: ",w2_0)
```

#Finding the approximation

```python
#Step-4
for i in range(1,n):
    x = a+(i*h)
    print("x: ",x)
    W_1 = u1[i]+(w2_0*v1[i])
    print("W_1 : ",W_1)
#    W_2 = u2[i]+(w2_0*v2[i])
#    print("W_2 : ",W_2)
```
